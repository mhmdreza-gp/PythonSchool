```python
import ee
#
# FeatureCollection پلی‌گون‌ها
region_fc = ee.FeatureCollection(region)

# پارامترها
target_year = 2016      # یک سال دلخواه (برای NDSI استاندارد >= 1984)
month = 7               # تیر
ndsi_threshold = 0.4    # آستانه NDSI

# انتخاب کالکشن و نام باندها بر اساس سال
if target_year >= 2021:
    col_id = 'LANDSAT/LC09/C02/T1_L2'; green_band, swir1_band = 'SR_B3', 'SR_B6'; use_cirrus = True
elif target_year >= 2013:
    col_id = 'LANDSAT/LC08/C02/T1_L2'; green_band, swir1_band = 'SR_B3', 'SR_B6'; use_cirrus = True
elif target_year >= 1999:
    col_id = 'LANDSAT/LE07/C02/T1_L2'; green_band, swir1_band = 'SR_B2', 'SR_B5'; use_cirrus = False
elif target_year >= 1984:
    col_id = 'LANDSAT/LT05/C02/T1_L2'; green_band, swir1_band = 'SR_B2', 'SR_B5'; use_cirrus = False
else:
    raise ValueError('برای NDSI به SWIR1 نیاز داریم؛ سال باید >= 1984 باشد.')

start = ee.Date.fromYMD(target_year, month, 1)
end   = start.advance(1, 'month')

# ماسک ابر/سایه‌ابر (QA_PIXEL)
def mask_clouds_shadows(img):
    qa = img.select('QA_PIXEL')
    dilated = qa.bitwiseAnd(1 << 1).neq(0)
    cloud   = qa.bitwiseAnd(1 << 3).neq(0)
    shadow  = qa.bitwiseAnd(1 << 4).neq(0)
    if use_cirrus:
        cirrus = qa.bitwiseAnd(1 << 2).neq(0)
        mask = (dilated.Or(cloud).Or(shadow).Or(cirrus)).Not()
    else:
        mask = (dilated.Or(cloud).Or(shadow)).Not()
    return img.updateMask(mask)

# افزودن NDSI روی رفلکتنس مقیاس‌خورده (C2 L2 scale/offset)
def add_ndsi(img):
    scale = 0.0000275
    offset = -0.2
    g  = img.select(green_band).multiply(scale).add(offset)
    s1 = img.select(swir1_band).multiply(scale).add(offset)
    ndsi = g.subtract(s1).divide(g.add(s1)).rename('NDSI')
    return img.addBands(ndsi)

# کالکشن تیر آن سال، فقط داخل پلی‌گون‌ها
col = (ee.ImageCollection(col_id)
       .filterDate(start, end)
       .filterBounds(region_fc.geometry())
       .map(mask_clouds_shadows)   # فقط ماسک ابر/سایه‌ابر طبق QA_PIXEL
       .map(add_ndsi))

print('image count:', col.size().getInfo())

# کامپوزیت و ماسک برف
composite = col.median()
snow = composite.select('NDSI').gt(ndsi_threshold)

# مساحت پیکسل (m²) فقط روی برف
area_img = ee.Image.pixelArea().updateMask(snow)

# مساحت برف برای هر پلی‌گون (بدون تعیین scale و maxPixels)
areas_fc = area_img.reduceRegions(
    collection = region_fc,
    reducer    = ee.Reducer.sum().setOutputs(['snow_m2'])
)

# چاپ نتیجه
result = areas_fc.getInfo()
for f in result['features']:
    p = f['properties']
    pid = p.get('id') or p.get('ID') or p.get('Id')
    print(f"Year {target_year} | Feature {pid}: snow area (m²) = {p.get('snow_m2')}")
```

```python
# --- خلاصه‌ی کیفیت نتایج روی پلیگون‌ها (همه در یک بلاک) ---

# 1) مساحت کل پلیگون و مساحت برف و درصد برف
poly_area = ee.Image.pixelArea().rename('poly_m2')
snow_area_img = ee.Image.pixelArea().updateMask(snow).rename('snow_m2')

stats_fc = ee.Image.cat([poly_area, snow_area_img]).reduceRegions(
    collection = region_fc,
    reducer    = ee.Reducer.sum()   # خروجی‌ها با نام باندها برمی‌گردند: poly_m2 و snow_m2
)

stats = stats_fc.getInfo()
print('--- Snow area vs polygon area ---')
for f in stats['features']:
    p = f['properties']
    pid = p.get('id') or p.get('ID') or p.get('Id')
    poly_km2 = (p.get('poly_m2') or 0)/1e6
    snow_km2 = (p.get('snow_m2') or 0)/1e6
    pct = (snow_km2/poly_km2*100) if poly_km2 else 0
    print(f"Feature {pid}: area={poly_km2:.1f} km² | snow={snow_km2:.1f} km² ({pct:.2f}%)")

# 2) مساحت ناحیه‌ی معتبر بعد از ماسک ابر/سایه (برای اطمینان از اینکه صفرها به‌خاطر ابری بودن نیست)
valid_mask = composite.select('NDSI').mask().gt(0)
valid_area_img = ee.Image.pixelArea().updateMask(valid_mask).rename('valid_m2')

valid_fc = valid_area_img.reduceRegions(
    collection = region_fc,
    reducer    = ee.Reducer.sum()   # خروجی: valid_m2
).getInfo()

print('\n--- Cloud-free (valid) area ---')
for f in valid_fc['features']:
    p = f['properties']
    pid = p.get('id') or p.get('ID') or p.get('Id')
    valid_km2 = (p.get('valid_m2') or 0)/1e6
    print(f"Feature {pid}: valid (cloud-free) area = {valid_km2:.1f} km²")

# 3) پرسن­tایل‌های NDSI داخل هر پلیگون (برای چک کردن آستانه 0.4)
def ndsi_percentiles(feat):
    d = composite.select('NDSI').reduceRegion(
        reducer = ee.Reducer.percentile([5, 50, 95]),
        geometry = feat.geometry()
    )
    return feat.set(d)

ndsi_ps_fc = region_fc.map(ndsi_percentiles).getInfo()

print('\n--- NDSI percentiles (p5, p50, p95) ---')
for f in ndsi_ps_fc['features']:
    p = f['properties']
    pid = p.get('id') or p.get('ID') or p.get('Id')
    print(f"Feature {pid}: NDSI p5={p.get('NDSI_p5')}, p50={p.get('NDSI_p50')}, p95={p.get('NDSI_p95')}")
```

```python
import ee
#

# پلی‌گون‌ها از shapefile
region_fc = ee.FeatureCollection(region)

# پارامترها
target_year    = 2016     # سال دلخواه (برای NDSI استاندارد >= 1984)
month          = 7        # تیر
ndsi_threshold = 0.4      # آستانه NDSI
cloud_max      = 40       # حداکثر درصد ابر براساس متادیتا

# انتخاب کالکشن و نام باندها
if target_year >= 2021:
    col_id = 'LANDSAT/LC09/C02/T1_L2'; G, S1 = 'SR_B3','SR_B6'
elif target_year >= 2013:
    col_id = 'LANDSAT/LC08/C02/T1_L2'; G, S1 = 'SR_B3','SR_B6'
elif target_year >= 1999:
    col_id = 'LANDSAT/LE07/C02/T1_L2'; G, S1 = 'SR_B2','SR_B5'
elif target_year >= 1984:
    col_id = 'LANDSAT/LT05/C02/T1_L2'; G, S1 = 'SR_B2','SR_B5'
else:
    raise ValueError('NDSI به SWIR1 نیاز دارد؛ سال باید ≥ 1984 باشد.')

start = ee.Date.fromYMD(target_year, month, 1)
end   = start.advance(1, 'month')

# افزودن NDSI روی رفلکتنسِ C2 L2 (scale/offset رسمی)
def add_ndsi(img):
    scale, offset = 0.0000275, -0.2
    g  = img.select(G).multiply(scale).add(offset)
    s1 = img.select(S1).multiply(scale).add(offset)
    ndsi = g.subtract(s1).divide(g.add(s1)).rename('NDSI')
    return img.addBands(ndsi)

# کالکشن اولیه
raw_col = (ee.ImageCollection(col_id)
           .filterDate(start, end)
           .filterBounds(region_fc.geometry()))

# فیلتر فقط بر اساس درصد ابری متادیتا (LAND اولویت دارد، اگر نبود از CLOUD_COVER)
cloud_filter = ee.Filter.Or(
    ee.Filter.lte('CLOUD_COVER_LAND', cloud_max),
    ee.Filter.lte('CLOUD_COVER',      cloud_max)
)
col = raw_col.filter(cloud_filter).map(add_ndsi)

# گزارش تعداد قبل/بعد از فیلتر و میانگین ابر متادیتا
print('total images (raw):', raw_col.size().getInfo())
print('filtered by cloud% :', col.size().getInfo())
print('mean CLOUD_COVER_LAND:', col.aggregate_mean('CLOUD_COVER_LAND').getInfo())
print('mean CLOUD_COVER     :', col.aggregate_mean('CLOUD_COVER').getInfo())

# کامپوزیت و ماسک برف
composite = col.median()
snow = composite.select('NDSI').gt(ndsi_threshold)

# مساحت برف برای هر پلی‌گون (m²) — بدون تعیین scale و maxPixels
snow_m2_img = ee.Image.pixelArea().updateMask(snow).rename('snow_m2')
areas_fc = snow_m2_img.reduceRegions(
    collection = region_fc,
    reducer    = ee.Reducer.sum().setOutputs(['snow_m2'])
).getInfo()

print('--- Snow area (m²) per feature ---')
for f in areas_fc['features']:
    p = f['properties']
    pid = p.get('id') or p.get('ID') or p.get('Id')
    print(f"Year {target_year} | Feature {pid}: snow_m2 = {p.get('snow_m2')}")
```

```python
import ee
#

# --- ورودی‌ها ---
region_fc = ee.FeatureCollection(region)   # پلی‌گون‌ها از shapefile
start_year     = 1984                      # NDSI استاندارد از 1984
end_year       = 2025
month          = 7                         # تیر
ndsi_threshold = 0.4
cloud_max      = 40                        # حداکثر درصد ابر متادیتا (LAND یا TOTAL)

# کالکشن‌های لندست را با نام‌بندی یکنواخت باندها (GREEN, SWIR1) آماده می‌کنیم
def prep(col_id, g, s1):
    return ee.ImageCollection(col_id).select([g, s1], ['GREEN', 'SWIR1'])

landsat_merged = (
    prep('LANDSAT/LT05/C02/T1_L2', 'SR_B2', 'SR_B5')  # L5 TM
    .merge(prep('LANDSAT/LE07/C02/T1_L2', 'SR_B2', 'SR_B5'))  # L7 ETM+
    .merge(prep('LANDSAT/LC08/C02/T1_L2', 'SR_B3', 'SR_B6'))  # L8 OLI
    .merge(prep('LANDSAT/LC09/C02/T1_L2', 'SR_B3', 'SR_B6'))  # L9 OLI-2
)

# فیلتر ابر فقط از متادیتا
cloud_filter = ee.Filter.Or(
    ee.Filter.lte('CLOUD_COVER_LAND', cloud_max),
    ee.Filter.lte('CLOUD_COVER',      cloud_max)
)

# افزودن NDSI روی Reflectance مقیاس‌خورده‌ی C2 L2
def add_ndsi(img):
    scale, offset = 0.0000275, -0.2
    g  = img.select('GREEN').multiply(scale).add(offset)
    s1 = img.select('SWIR1').multiply(scale).add(offset)
    ndsi = g.subtract(s1).divide(g.add(s1)).rename('NDSI')
    return img.addBands(ndsi)

# محاسبه برف سالانه (تیر) برای یک سال و برگرداندن FeatureCollection با پروپرتی year
def yearly_snow(y):
    y = ee.Number(y)
    start = ee.Date.fromYMD(y, month, 1)
    end   = start.advance(1, 'month')
    col_y = (landsat_merged
             .filterDate(start, end)
             .filterBounds(region_fc.geometry())
             .filter(cloud_filter)
             .map(add_ndsi))
    comp  = col_y.median()
    snow  = comp.select('NDSI').gt(ndsi_threshold)
    snow_area_img = ee.Image.pixelArea().updateMask(snow).rename('snow_m2')
    fc = snow_area_img.reduceRegions(
        collection = region_fc,
        reducer    = ee.Reducer.sum().setOutputs(['snow_m2'])
    )
    return fc.map(lambda f: f.set('year', y))

# اجرای سروری برای کل سال‌ها و گرفتن یک‌باره خروجی
years = ee.List.sequence(start_year, end_year)
fc_all = ee.FeatureCollection(years.map(yearly_snow)).flatten()

# چاپ نتایج (سال، شناسه پلی‌گون، مساحت برف m² و همچنین به km²)
res = fc_all.getInfo()
print('year, feature_id, snow_m2, snow_km2')
for f in res['features']:
    p = f['properties']
    y = int(round(p['year']))
    fid = p.get('id') or p.get('ID') or p.get('Id')
    snow_m2 = p.get('snow_m2') or 0
    print(f"{y}, {fid}, {snow_m2}, {snow_m2/1e6:.3f}")
```
